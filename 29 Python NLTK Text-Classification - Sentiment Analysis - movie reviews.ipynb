```python
import nltk
import random
from nltk.corpus import movie_reviews
```

```python
movie_reviews.categories()
```

```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append([movie_reviews.words(fileid), category])

len(documents)
```

```python
documents[1200]
```

```python
random.shuffle(documents)
```

```python
documents[1200]
```

```python
from nltk.corpus import stopwords
stop = stopwords.words("english")
```

```python
stop2 = [',', '.', "'", '"', '-', ')', '(', ':', '?']
words = []
for word in movie_reviews.words():
    if word not in stop and word not in stop2:
        words.append(word.lower())

words_FD = nltk.FreqDist(words)
```

```python
words_FD.most_common(10)
```

```python
len(words)
```

```python
words_FD['good']
```

```python
word_features = list(words_FD.keys())[:10000]
```

```python
len(documents)
```

```python
documents[0]
```

```python
featuresets = []
for (review, category) in documents:
    words = set(review)
    features = {}
    for w in word_features:
        features[w] = w in words
    featuresets.append([features, category])
len(featuresets)
```

```python
featuresets[0]
```

```python
training_set = featuresets[:1800] # 1800 
testing_set = featuresets[1800:] # 200
```

```python
clf = nltk.NaiveBayesClassifier.train(training_set)
```

```python
nltk.classify.accuracy(clf, testing_set)
```
