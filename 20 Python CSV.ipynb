```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import itertools
import csv
```

```python
files = ['Restaurant.csv'] # polarity,text,restaurant_id,user_id
```

```python
def prepareDatasets(files):
    sentences = []
    
    for file_name in files:
        file = open(file_name, "rt", encoding="utf-8")
        reviews = csv.reader(file)
        for polarity,text,restaurant_id,user_id in reviews:
            sentences.append([text, polarity])
        
    df = DataFrame(sentences, columns=['text', 'polarity'])
    return df
```

```python
reviews = prepareDatasets(files)
```

```python
import mysql.connector
from mysql.connector import errorcode, connect
```

```python
def connectMysql():
    try:
        cnn = connect(
            user='root', 
            password='root', 
            host='localhost', 
            database='test',
            unix_socket='/Applications/MAMP/tmp/mysql/mysql.sock')
        return cnn
    except mysql.connector.Error as e:
        if e.errno == errorcode.ER_ACCESS_DENIED_ERROR:
            print ("Access denied!")
        elif e.errno == errorcode.ER_BAD_DB_ERROR:
            print ("Database does not exist.")
        else:
            print (e)
```

```python
def createTable(sql):
    try:
        cnn = connectMysql()
        cursor = cnn.cursor()
        cursor.execute(sql)
    except:
        cnn.rollback()
    cnn.close()
```

```python
sql = "DROP TABLE IF EXISTS REVIEWS; CREATE TABLE REVIEWS (TEXT TEXT NOT NULL, POLRITY INT (10))"
createTable(sql)
```

```python
def insertData(text, polarity):
    try:
        cnn = connectMysql()
        cursor = cnn.cursor()
        cursor.execute("INSERT INTO `REVIEWS` VALUES ('%s', '%s')" % (text, polarity))
        cnn.commit()
    except:
        cnn.rollback()
    cnn.close()
```

```python
def getData(sql):
    cnn = connectMysql()
    cursor = cnn.cursor()
    cursor.execute(sql)
    reviews = DataFrame(cursor.fetchall())
    reviews.rename(columns={0:'text', 1:'polarity'}, inplace=True)
    cursor.close()
    cnn.close()
    return reviews
```

```python
sql ="SELECT * FROM `REVIEWS`"
getData(sql)
```
