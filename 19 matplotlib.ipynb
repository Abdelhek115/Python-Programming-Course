```python
emps = [1,2,3,4,5]

task_a = [20, 10, 40, 30, 35]
task_b = [20, 30, 20, 30, 30]
task_c = [35, 10, 20, 20, 20]
task_d = [25, 50, 20, 20, 15]

plt.plot([],[], color='c', label='task_a', linewidth=5)
plt.plot([],[], color='g', label='task_b', linewidth=5)
plt.plot([],[], color='wheat', label='task_c', linewidth=5)
plt.plot([],[], color='y', label='task_d', linewidth=5)

plt.stackplot(emps, task_a, task_b, task_c, task_d, colors=['c', 'g', 'wheat', 'y'])

plt.xlabel("Employee #")
plt.ylabel('Work %')
plt.legend()
plt.title('Employees Works')
```

```python
slices = [20, 15, 30, 25]
tasks = ['task_a', 'task_b', 'task_c', 'task_d']
cols = ['wheat','c','royalblue','y']

plt.pie(slices,
        labels=tasks,
        colors=cols,
        shadow=False,
        startangle=190,
        explode=(0,0,0, .05),
        autopct='%1.1f%%'
       )
plt.title("Pie Chart title")
```
